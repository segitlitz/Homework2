```python
data_2014["year"] = 2014
data_2015["year"] = 2015
data_2016["year"] = 2016
data_2017["year"] = 2017
data_2018["year"] = 2018
data_2019["year"] = 2019
```

```python
data = pd.concat([data_2014, data_2015, data_2016, data_2017, data_2018, data_2019], ignore_index=True)
```

```python
data.columns
```

## Question 1

```python
question_1 = data[
    (data["snp"] != "Yes") &
    ~((data["planid"] >= 800) & (data["planid"] < 900)) & 
    ~ (data["payment_partc"].isna())].copy()
question_1.head(10)
```

```python
plan_counts = (
    question_1
    .groupby(["year", "county"])["planid"]
    .nunique()
    .reset_index(name="num_unique_plans")
)
plan_counts.head()
```

```python
plan_counts.boxplot(column="num_unique_plans", by="year")
plt.title("Plans Counts by County")
plt.suptitle("")  # removes automatic pandas title
plt.ylabel("Number of Plans in County")
plt.show()
```

I believe the number of plans by county is sufficient and is increasing each year, showing increasing competition.

## Question 3

```python
question_1.info()
```

```python
# HHI = sum (avg_enrolled/county_enrolled)^2
question_1["market_share"] = (
    question_1["avg_enrolled"] /
    question_1.groupby(["year", "county"])["avg_enrolled"].transform("sum")
)

question_1["HHI"] = question_1["market_share"] ** 2

question_1["HHI_sum"] = (
    question_1
    .groupby(["year", "county"])["HHI"]
    .transform("sum")
)
question_1.head()
```

```python
average_HHI = (
    question_1
    .groupby("year", as_index=False)["HHI_sum"]
    .mean()
)
average_HHI.head()
```

```python
# Plot HHI Overtime
plt.figure()
plt.plot(average_HHI["year"], average_HHI["HHI_sum"])
plt.xlabel("Year")
plt.ylabel("Average HHI")
plt.title("Average HHI Over Time")
plt.show()
```

The average HHI decreases substantially overtime. 

## Question 4

```python
# Average Share of Medicare Advantage = sum avg_enrolled / avg_eligible 
county_MA = (
    question_1
    .groupby(["year", "county"], as_index=False)
    .agg(
        enrolled=("avg_enrolled", "sum"),
        eligibles=("avg_eligibles", "first")
    )
)

county_MA["share_MA"] = county_MA["enrolled"] / county_MA["eligibles"]

average_share_MA = (
    county_MA
    .groupby("year", as_index=False)["share_MA"]
    .mean()
)
average_share_MA
```

```python
# Plot Average Share MA
plt.figure()
plt.plot(average_share_MA["year"], average_share_MA["share_MA"])
plt.xlabel("Year")
plt.ylabel("Average Share MA")
plt.title("Average Share MA over time")
plt.show()
```
